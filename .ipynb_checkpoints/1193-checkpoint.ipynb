# 분수찾기

### 문제
무한히 큰 배열에 다음과 같이 분수들이 적혀있다.
  
1/1|1/2|1/3|1/4|1/5|…  
2/1|2/2|2/3|2/4|…	…  
3/1|3/2|3/3|…	…	…  
4/1|4/2|…	…	…	…  
5/1|…	…	…	…	…  
…	…	…	…	…	…  

이와 같이 나열된 분수들을 1/1 -> 1/2 -> 2/1 -> 3/1 -> 2/2 -> … 과 같은 지그재그 순서로 차례대로 1번, 2번, 3번, 4번, 5번, … 분수라고 하자.  
  
X가 주어졌을 때, X번째 분수를 구하는 프로그램을 작성하시오.  
  
### 입력
첫째 줄에 X(1 ≤ X ≤ 10,000,000)가 주어진다.  
  
### 출력
첫째 줄에 분수를 출력한다.  
  
### 예제 입력 1  
14  
### 예제 출력 1 
2/4

```python
N = int(input())

numer = 1
denom = 2
orientation = 0 #0: left and down, 1: right and up, 2: else

for i in range(N-2):
    
    if orientation == 0 and denom != 1:
        numer += 1
        denom -= 1
    
    elif orientation == 1 and numer != 1:
        numer -= 1
        denom += 1
    
    elif numer == 1 and denom != 1:
        denom += 1
        orientation = 0
        
    elif denom == 1 and numer != 1:
        numer += 1
        orientation = 1
        
print('{}/{}' .format(numer, denom))
```

```python
N = int(input())

numer = 1
denom = 2
orientation = 0 #0: left and down, 1: right and up, 2: else
print()
for i in range(N-2):
    
    if orientation == 0 and denom != 1:
        numer += 1
        denom -= 1
    
    elif orientation == 1 and numer != 1:
        numer -= 1
        denom += 1
    
    elif numer == 1 and denom != 1:
        denom += 1
        orientation = 0
        print('--------')
    elif denom == 1 and numer != 1:
        numer += 1
        orientation = 1
        print('========')
        
    print('{}: {}/{}' .format(i+3,numer, denom))
# print('{}/{}' .format(numer, denom))
```
